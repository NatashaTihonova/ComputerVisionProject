```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
print(cv2.__version__)
```

# Задание 1.

```python
def showImage(imageForShowing):
    plt.figure(figsize = (10, 10))
    plt.imshow(imageForShowing)
```

```python
name_img_blood = 'cars.png'
img_blood = cv2.imread(name_img_blood, cv2.IMREAD_GRAYSCALE)
original = cv2.imread(name_img_blood, cv2.IMREAD_GRAYSCALE)
showImage(img_blood)
```

```python
blur_img_blood = cv2.medianBlur(img_blood, 9)
#blur_img_blood = cv2.bilateralFilter(img_blood, 50, 50, cv2.BORDER_REFLECT)
showImage(blur_img_blood);
```

```python
ret, binary_img_blood = cv2.threshold(blur_img_blood,150,255,cv2.THRESH_BINARY)
print(ret)
showImage(binary_img_blood);
```

```python
contours, hierarchy = cv2.findContours(binary_img_blood, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_TC89_L1)
cv2.drawContours(img_blood, contours, -1, (0,0,0), 3, cv2.LINE_8, hierarchy, 1)
showImage(img_blood)
print("Количество клеток: ", int(len(contours)))
```

```python
plt.figure(figsize = (120, 120))
plt.subplot(10, 10, 1)
plt.imshow(original)
plt.subplot(10, 10, 2)
plt.imshow(img_blood)
plt.show()
```

# Задание 2.

```python
name_img_sugar = "Sugars.png"
sugars_spoons_img = cv2.imread(name_img_sugar, cv2.IMREAD_GRAYSCALE)
original_sugars_spoons = cv2.imread(name_img_sugar, cv2.IMREAD_GRAYSCALE)
```

```python
showImage(sugars_spoons_img)
```

```python
blur_sugar_spoons_img = cv2.medianBlur(sugars_spoons_img, 15)
showImage(blur_sugar_spoons_img)
```

```python
ret, binary_sugar_spoons_img = cv2.threshold(blur_sugar_spoons_img, 220, 255, cv2.THRESH_BINARY)
showImage(binary_sugar_spoons_img)
```

```python
contours, hierarchy = cv2.findContours(binary_sugar_spoons_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_TC89_L1)
cv2.drawContours(sugars_spoons_img, contours, -1, (0,0,0), 3, cv2.LINE_8, hierarchy, 1)
for contour in contours:
    if (cv2.contourArea(contour) < 200):
        contours.remove(contour)
showImage(sugars_spoons_img)
print("Количество объектов: ", int(len(contours))) 
#Десятый объект само изображение
```

## Поиск сахара

```python
#Параметры SimpleBlobDetector для сахара.
params_sugar = cv2.SimpleBlobDetector_Params()
# Фильтр по цветуу - ищем только белые пиксели
params_sugar.filterByColor = True
params_sugar.blobColor = 255
# Фильтр по выпуклости - ищем только выпуклые объекты
params_sugar.filterByConvexity = True
params_sugar.minConvexity = 0.5
# Фильтр по развернутости изображения - ищем только развернутые к нам изображения
params_sugar.filterByInertia = True
params_sugar.minInertiaRatio = 0.5
detector_sugars = cv2.SimpleBlobDetector_create(params_sugar)
sugars_keypoints = detector_sugars.detect(binary_sugar_spoons_img)
sugars_with_keypoints = cv2.drawKeypoints(sugars_spoons_img,sugars_keypoints,np.array([]),(255,0,0),
                                          cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
showImage(sugars_with_keypoints)
print("Количество сахара: ", len(sugars_keypoints))
```

## Поиск ложек

```python
#Параметры SimpleBlobDetector для всех объектов.
params_objects = cv2.SimpleBlobDetector_Params()
# Фильтр по цвету - ищем только белые пиксели
params_objects.filterByColor = True
params_objects.blobColor = 255

detector_objects = cv2.SimpleBlobDetector_create(params_objects)
objects_keypoints = detector_objects.detect(binary_sugar_spoons_img)
objects_with_keypoints = cv2.drawKeypoints(sugars_spoons_img,objects_keypoints,np.array([]),(0,255,0),
                                          cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

print("Количество найденных объектов : ", len(objects_keypoints))
print("Количество контуров (объектов): ", len(contours))

showImage(objects_with_keypoints)
```

```python
if (len(objects_keypoints) == len(contours)):
    print("Все объекты найдены")
else:
    detected_spoons = cv2.imread(name_img_sugar, cv2.IMREAD_GRAYSCALE)
    listContours = []
    listContours.extend(contours)
    for contour in range(len(contours)):
        fill_original = cv2.imread(name_img_sugar, cv2.IMREAD_GRAYSCALE)
        cv2.drawContours(fill_original, contours, contour, (0,0,0) ,cv2.FILLED)
        for keypoint in objects_keypoints:
          x = int(keypoint.pt[0])
          y = int(keypoint.pt[1])
          pixel = fill_original[y, x]
          if (pixel == 0):
            listContours.remove(contours[contour])
    cv2.drawContours(detected_spoons, listContours, -1, (0,0,0), cv2.LINE_8)
    showImage(detected_spoons)
    print("Количество ложек: ", len(listContours))
```

```python
plt.figure(figsize = (120, 120))
plt.subplot(10, 10, 1)
plt.imshow(sugars_spoons_img)
plt.subplot(10, 10, 2)
plt.imshow(sugars_with_keypoints)
plt.subplot(10, 10, 3)
plt.imshow(detected_spoons)
plt.show()
```

# Задание 4.

```python
name_numbers_cars = 'cars.png'
img_numbers_cars = cv2.imread(name_numbers_cars, cv2.IMREAD_GRAYSCALE)
showImage(img_numbers_cars)
```

```python
blur_image_cars = cv2.medianBlur(img_numbers_cars, 3)
showImage(blur_image_cars)
```

```python
edged = cv2.Canny(blur_image_cars, 170, 255)
showImage(edged)
```

```python
contours_objects, hierarchy = cv2.findContours(edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_TC89_L1)
contours_objects = sorted(contours_objects, key = cv2.contourArea, reverse = True) [:30]
```

```python
NumberPlateCnt = None
original_img_number = cv2.imread(name_numbers_cars)
for contour in contours_objects:
        peri = cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, 0.02 * peri, True)
        if len(approx) == 4:
            NumberPlateCnt = approx
            original_img_number = cv2.drawContours(original_img_number, [NumberPlateCnt], -1, (255, 0, 255), cv2.LINE_4)
```

```python
showImage(original_img_number)
```
