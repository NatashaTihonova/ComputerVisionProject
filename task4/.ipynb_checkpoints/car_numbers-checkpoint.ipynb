```python
import pylab as plt
import numpy as np
import matplotlib.pyplot as pplt
import cv2
import math
import imutils
%matplotlib inline

image = cv2.imread('cars.png')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

f, axarr = plt.subplots(1, 1, figsize = (25, 25))
axarr.imshow(image, cmap="gray")
plt.show()

def rotate_image(image, angle):
  image_center = tuple(np.array(image.shape[1::-1]) / 2)
  rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
  result = cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
  return result

#каскадное обнаружение объектов Хаара
license_plate_cascade = cv2.CascadeClassifier('haarcascade_russian_plate_number.xml')
```

```python
rot_image = gray_image.copy()
rot_image = imutils.rotate_bound(image, 45)
rot_image = imutils.rotate_bound(rot_image, -45)
rot_image = cv2.cvtColor(rot_image, cv2.COLOR_BGR2GRAY)
count = 0;

rect_centers = []
angle = 45

for i in range(0, 3):
    license_plates = license_plate_cascade.detectMultiScale(rot_image)
 
    for (x, y, w, h) in license_plates:
        if (w > 2 * h):
            start_point = (x, y)
            end_point = (x + w, y + h)
            
            cross = False
            for c in rect_centers:
                if (c[0] >= start_point[0] and c[1] >= start_point[1] and c[0] <= end_point[0] and c[1] <= end_point[1]):
                    cross = True
            if (cross == True):
                continue
                
            center = (start_point[0] + int(w * 0.8)/2, start_point[1] + int(h * 0.7)/2)
            rect_centers.append(center)
            
            cv2.rectangle(rot_image, start_point, end_point, (255,0,0), 2)
            count += 1
            
    f, axarr = plt.subplots(1, 1, figsize = (15, 15))
    axarr.imshow(rot_image, cmap="gray")
    plt.show()
    
    rot_image = rotate_image(rot_image, angle)
    angle = -90
        
print('Количество номеров: ', count)

rot_image = rotate_image(rot_image, 135)
f, axarr = plt.subplots(1, 1, figsize = (25, 25))
axarr.imshow(rot_image, cmap="gray")

plt.show()
```
